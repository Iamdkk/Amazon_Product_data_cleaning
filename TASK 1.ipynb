### Removing Duplicates

```python
df.drop_duplicates(inplace=True)
```

### Fixing Null values in rating count

```python
df['rating_count'] = df['rating_count'].fillna('0')
```

```python
df['rating_count'].isnull()
```

### Fixing Null values in rating

```python
df['rating'] = df['rating'].fillna(0)
```

```python
df.isnull().sum()
```

### Removing Symbols and Commas

```python
def clean_price(price):
    return float(price.replace('₹', '').replace(',', '').strip()) if isinstance(price, str) else price
```

```python
def clean_discount(discount):
    return float(discount.replace('%', '').strip()) if isinstance(discount, str) else discount
```

```python
def clean_rating_count(count):
    return int(count.replace(',', '').strip()) if isinstance(count, str) else count
```

### Cleaning the prices,rating and count

```python
df['discounted_price'] = df['discounted_price'].apply(clean_price)
df['actual_price'] = df['actual_price'].apply(clean_price)
df['discount_percentage'] = df['discount_percentage'].apply(clean_discount)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = df['rating_count'].apply(clean_rating_count)
```

### Reseting Indexing

```python
df.reset_index(drop=True, inplace=True)
```
